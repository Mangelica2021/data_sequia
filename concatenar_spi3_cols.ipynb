# Concatenar los archivos de las estaciones en uno solo

```python
import pandas as pd
import numpy as np
```

```python
'''leer DF con los nombres de los archivos'''
nombres = pd.read_csv('stations_temp.txt', header = None)
```

```python
'''ciclo for para leer los datos'''
df_concatenar = pd.DataFrame(columns=["fecha"])
for i in range(nombres.shape[0]):
    '''ruta de cada archivo a leer'''
    cpath01 = 'estaciones_spi/indices_'
    cnombre01 = str(nombres.iloc[i,4])
    cfile01 = cpath01 + cnombre01 + '.csv'
    df_aux = pd.read_csv(cfile01)
    df_aux1= pd.DataFrame()
    
#     '''columna fecha'''    
#     df_concatenar['fecha'] = df_concatenar['anio'].astype(str) + '-' + df_concatenar['mes'].astype(str) + '-1'
# #     '''cambiar orden de las fechas'''
# #     df_concatenar = df_concatenar[['fecha', 'codigo', 'pp', 'tmax', 'tmin', 'spi3']]

    '''columna con el codigo de la estacion'''
    nom_cols = cnombre01
    df_aux1[nom_cols] = df_aux['spi3']
    '''concatenar el resuldado'''
    df_concatenar = pd.concat([df_concatenar, df_aux1], axis = 1)
    df_concatenar.reset_index(drop = True, inplace = True)

'''columna fecha'''    
df_concatenar['fecha'] = df_aux['anio'].astype(str) + '-' + df_aux['mes'].astype(str) + '-1'


'''grabar el DF resultado'''
cfile00 = 'estaciones_spi/sequias_spi_cols.csv'
df_concatenar.to_csv(cfile00, index = False)

    
```
